```python
df = df[df.columns[df.isna().sum()/df.shape[0] < 0.9]] # elimnation  des   colonnes unitiles jusqu'a 90%
df.head()
```

```python
df["NOMBRE_DE_CARTONS"]=df["QUANTITY"]
```

```python
for i in range (len(df['DESCRIPTION'])):
    num=''
    if not df['DESCRIPTION'][i][0].isdigit():
        num='1'
        df['NOMBRE_DE_CARTONS'][i]=float(num)
        continue
        
    for c in df['DESCRIPTION'][i]:
        if c.isdigit():
            num +=c
        else:
            break
    df['NOMBRE_DE_CARTONS'][i]=float(num)
     
    
```
